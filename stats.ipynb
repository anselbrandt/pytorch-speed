```python
file = "../data/train.txt"

speeds = []

with open(file) as f:
    for line in f: # read rest of lines
        speeds.append(float(line))

print(len(speeds))
```

```python
min_speed = min(speeds)
max_speed = max(speeds)

print(f"min {min_speed}\nmax {max_speed}")
```

```python
import matplotlib.pyplot as plt

plt.plot(speeds)
plt.xlabel('Frame Number')
plt.ylabel('Speed m/s')
plt.show()
```

```python
# greatest speed difference between frames

def maximumGap(nums: list[int]) -> int:
    max = 0
    values = ()
    for i in range(len(nums) - 1):
        diff = abs(nums[i] - nums[i + 1])
        if diff > max:
            max = diff
            values = (i,i+1)
    return values
    
print(maximumGap(speeds))
```

```python
from numpy import diff
import matplotlib.pyplot as plt

accel = diff(speeds)

plt.plot(accel)
plt.xlabel('Frame Number')
plt.ylabel('Accel m/s^2')
plt.show()
```

```python
# greatest acceleration
    
print(accel.tolist().index(max(accel)))
```
